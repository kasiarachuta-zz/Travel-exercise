```python
import pandas as pd
import numpy as np
```

```python
# I import the file with data

flights = pd.read_csv("BOS_CUN_trips1M.csv",)

flights.head()
```

```python
# I find the mean price of a flight for different departure and return dates combinations.

best_dates = flights.groupby(['departure_odate', 'return_ddate']).mean()

# Because I'm interested in only the total price, I can just pick this column.
# I then export it to CSV so I can visualise the data in Tableau.
best_dates['total'].to_csv("bestdates.csv")

best_dates['total']

# The visusalisation turns out to be very big and doesn't tell me much. I decided to explore the data further.
```

```python
# I find the mean price for the different date combinations.
# I then just pick the 'total' columns as that's the only one I'm interested in.

departurebest = flights.groupby(['departure_odate', 'return_ddate']).mean()
departurebesttotal = departurebest['total']

# I then look at some statistics for the different date combinations.
departurebesttotal.describe()

# I notice that the standard deviation is pretty high at $155
```

```python
# I now want to look at the date combinations below first quartile (cheapest 25% of flights)

# I want to turn this into pandas DataFrame object.
departurebesttotaldf = pd.DataFrame(departurebesttotal)
departurebesttotaldf

departurebesttotaldf[departurebesttotaldf['total'] <= 426.482274].to_csv("lowestquartileprice.csv")

departurebesttotaldf[departurebesttotaldf['total'] <= 426.482274]
# I have 2171 rows Ã— 1 columns.
# This means that there are 2171 different date combinations that are great to travel on.
# I spit these data out to visualise them in Tableau.
```

```python
# I want to look at the minimum price and when this happens.
# The minimum is ~363, so I'm going to look at date combinations below 364.

departurebesttotaldf[departurebesttotaldf['total'] <= 364]
```

```python
# I then want to explore date combinations below $400.

departurebesttotaldf[departurebesttotaldf['total'] <= 400]

# This gives me 1121 date combinations.
```

```python
# Let's try dates below $390.

departurebesttotaldf[departurebesttotaldf['total'] <= 390]

# 529 different date combinations.
```

```python
# I now want to to sort the difference prices.

lowestquartileprice = pd.read_csv("lowestquartileprice.csv")

lowestquartileprice.sort(columns='total')

# I now have sorted the dataframe by the total value so I get the cheapest departure and return date combinations.
```
