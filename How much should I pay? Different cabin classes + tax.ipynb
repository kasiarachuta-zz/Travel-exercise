```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Firstly, I read the data.

flights = pd.read_csv("BOS_CUN_trips1M.csv",)

flights.head()
```

```python
# I then add a column with the final total = total + tax

flights['final total'] = flights['total'] + flights['tax']

flights.head()
```

```python
# I first look at the economy class.

economy = flights[flights['cabinclass'] == 'E']

economy.head()
```

```python
# I now want to visualise the economy data.

sns.boxplot(economy['final total'], vert=False)
```

```python
'''From the box plot above, I can see how the data are spread out. I now want to find the exact numbers for 
the mean, quartiles etc.
Again, I am surprised to see that 25% of the tickets are above $1,750.'''

economytotal = economy['final total'].describe()

economytotal = pd.DataFrame(economytotal)

economytotal
```

```python
'''I can see that for an economy flight, we won't pay more than ~$4454, with the mean value being $795
and median being $708.
The standard deviation is high at $273.'''
```

```python
# I now want to look at business class flights

business = flights[flights['cabinclass'] == 'B']

business.head()

# 1041 rows × 31 columns
```

```python
# As above, I want to use a box plot to visualise the data.

sns.boxplot(business['final total'], vert=False)
```

```python
'''As in the previous visualisation (without tax), the data are quite spread out.
Again, I can see that there is a relatively big difference between the median
and the fourth quartile.'''

businesstotal = business['final total'].describe()

businesstotal = pd.DataFrame(businesstotal)

businesstotal
```

```python
'''For a business flight, the mean flight price is $902 and the median is $772

and I shouldnt expect to pay more than $1769.

Again, the standard deviation is marginally lower than before at $242.
I am surprised that the maximum price is $1768 - it seems relatively low to me. Same goes for the mean and
the median, I was expecting them to be much higher.'''
```

```python
# Finally, I look at first class flights.

first = flights[flights['cabinclass'] == 'F']

first.head()

# 276 rows × 31 columns
```

```python
# As above, I want to use a box plot to visualise the data.

sns.boxplot(first['final total'], vert=False)
```

```python
'''Again, I can see the distribution, which is much less spread out that the distribution for the business tickets.
I want to look at the quick stats and find the actual numbers for the mean and quartiles.'''

firsttotal = first['final total'].describe()

firsttotal = pd.DataFrame(firsttotal)

firsttotal
```

```python
'''On average, I should be looking at paying $848, with the median being $797.

I shouldnt pay more than $1550 per flight. Standard deviation is lowest at only $159 - which
means that my observation that the data is the least spread out compared to other tickets is true'''
```
