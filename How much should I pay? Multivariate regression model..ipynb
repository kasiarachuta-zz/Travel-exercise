```python
import pandas as pd
import statsmodels.api as sm
```

```python
# I import the file with data

flights = pd.read_csv("BOS_CUN_trips1M.csv",)

flights.head()
```

```python
# I now want to find the day of the week of departure.

flights['departure_odate'] = pd.to_datetime(flights['departure_odate'])
flights['day_of_week_dep'] = flights['departure_odate'].dt.dayofweek

days = {0:'Mon',1:'Tue',2:'Wed',3:'Thur',4:'Fri',5:'Sat',6:'Sun'}

flights['day_of_week_dep'] = flights['day_of_week_dep'].apply(lambda x: days[x])

flights.head()
```

```python
# I now find day of week for return date.

flights['return_ddate'] = pd.to_datetime(flights['return_ddate'])
flights['day_of_week_ret'] = flights['return_ddate'].dt.dayofweek

days = {0:'Mon',1:'Tue',2:'Wed',3:'Thur',4:'Fri',5:'Sat',6:'Sun'}

flights['day_of_week_ret'] = flights['day_of_week_ret'].apply(lambda x: days[x])

flights.head()
```

```python
'''Now I want to write my model. This is a multivariate regressiom model that predicts the price based on
a number of factors such as:
cabin class (cabinclass), 
number of outbound stops (outboundstops), 
number of return stops (returnstops),
airline travelled with (majorcarrieid),
day of the week travelled on (day_of_week_dep), 
day of the week returned on (day_of_week_ret),
whether it's a refundable or a non-refundable flight (refundable),
date of departure (departure_odate),
date of return (return_ddate)
difference betweeen the departure and return (los2)
source (source)
ticketing authority (merchant)
number of seats available (availableseats).

I exclude the number of days before booked as I know this relationship is not linear (see - polynomial regression)

I will use least squares method.
I need to make sure that I all of my variables are categorical variables.'''

flights['cabinclass'] = pd.Categorical(flights.cabinclass).codes
flights['majorcarrierid'] = pd.Categorical(flights.majorcarrierid).codes
flights['day_of_week_dep'] = pd.Categorical(flights.day_of_week_dep).codes
flights['day_of_week_ret'] = pd.Categorical(flights.day_of_week_ret).codes
flights['return_ddate'] = pd.Categorical(flights.return_ddate).codes
flights['departure_odate'] = pd.Categorical(flights.departure_odate).codes
flights['refundable'] = pd.Categorical(flights.refundable).codes
flights['source'] = pd.Categorical(flights.source).codes
flights['merchant'] = pd.Categorical(flights.merchant).codes

X = flights[['cabinclass', 'outboundstops', 'returnstops', 'majorcarrierid',
            'day_of_week_dep', 'day_of_week_ret', 'return_ddate', 'departure_odate', 'refundable', 'los2',
            'source', 'availableseats', 'merchant']]
y = flights[['total']]

X1 = sm.add_constant(X)
est = sm.OLS(y, X1).fit()

est.summary()
```

```python
'''This is a fairly terrible model - R squared value is only 0.137.
It seems like whether the source is refundable or not has a big influence on the total price,
so does the cabin class and source. Two other big factors are number of outbound and return stops.
P values are all small (which validates the model) and t values are large (again, implying significance).
Day of the week for departure and return doesnt make a big difference.'''
```
