```python
import pandas as pd
import numpy as np
```

```python
# I import the file with data

flights = pd.read_csv("BOS_CUN_trips1M.csv",)

flights.head()
```

```python
# I want to look at refundable flights first.

refundableflights = flights[flights['refundable'] == True]
refundableflights.head()
```

```python
# I now run quick statistics on the

refundableflights['total'].describe()
```

```python
# Now I want to look at the non-refundable flights.

nrefundableflights = flights[flights['refundable'] == False]
nrefundableflights.head()
```

```python
nrefundableflights['total'].describe()
```
