```python
import pandas as pd
import numpy as np
```

```python
# I import the file with data

flights = pd.read_csv("BOS_CUN_trips1M.csv",)

flights.head()
```

```python
# I first look at the economy class.

economy = flights[flights['cabinclass'] == 'E']

economy.head()

# 1201004 rows Ã— 31 columns
```

```python
# I then find quick stats of the economy flights. I am only interested in the 'total' column.

economy.describe()
```

```python
'''I can see that for an economy flight, we won't pay more than ~$4278, with the mean value being $673 
and median being $590''' 
```

```python
# I now want to look at business class flights

business = flights[flights['cabinclass'] == 'B']

business.head()

# 1041 rows Ã— 31 columns
```

```python
# Again, I run quick statistics.

business.describe()
```

```python
'''For a business flight, the mean flight price is $786 and the median is $659 so I should expect to pay something around that'''
```
