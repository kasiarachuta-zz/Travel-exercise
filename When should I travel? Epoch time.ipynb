```python
import pandas as pd
import numpy as np
import time
import datetime
```

```python
# I import the file with data

flights = pd.read_csv("BOS_CUN_trips1M.csv",)

flights.head()
```

```python
# The departure and return columns are in epoch-seconds.
# I now want to look at which epoch seconds combinations for departure and return dates are cheapest.
# I am taking the average for each departure and return time/date combination

cheapest = flights.groupby(['departure', 'return']).mean()

# I only need the 'total' column with the total price.
cheapest = cheapest['total']

cheapestdf = pd.DataFrame(cheapest)

cheapestdf.describe()
```

```python
# I now want to look at the comvinations in the first quartile.

lowestquartile = cheapestdf[cheapestdf['total'] <= 452]

lowestquartile

# 60264 rows Ã— 1 columns so I still get too many combinations.
```
